El sistema de registro de contenedores del “Manzanillo International Terminal (MIT)” requiere un programa que organice los contenedores de las  compañías que desean importar mercancía a Panamá con un máximo de 30 contenedores. Si una reserva ya está siendo ocupada por una compañía, está deberá ser reubicada en el registro, mostrando así su posición inicial en el registro y su posición reubicada. Se les permitirá a las compañías verificar su posición de reserva del registro y su clave única ingresando el nombre de la compañía

```python
import time
import os
from IPython.display import clear_output


class hash_table:
  def __init__(self, tableSize):
    self.tableSize = tableSize
    self.table = [None] * tableSize

  # Función hash
  def funcion_hash(self, frase):
    key = 0
    for i in range(0,len(frase)):
      key += ord(frase[i])
    return key % self.tableSize

  def insertar(self, frase): # Metodo para ingresar elementos
    index = self.funcion_hash(frase)
    if self.table[index] is None:
      self.table[index] = frase
    elif index > self.tableSize: 
      print(f"Lo sentimos {frase} el puesto {index+1} ya esta ocupado.")
      self.insertararColision(frase, 0, 0)
    else:
      print(f"Lo sentimos {frase} el puesto {index+1} ya esta ocupado.")
      self.insertararColision(frase, 0, 0)

  def insertararColision(self, frase, index, limite):
        if limite < self.tableSize:
            if self.table[index] is None:
                print(f"Su nuevo puesto es el {index+1}.")
                self.table[index] = frase
            else:
                self.insertararColision(frase, index+1, limite+1)
        else:
            print(f"Lo sentimos {frase} los {self.tableSize} espacios estan ocupados")

  def buscar(self,value): # Metodo para retornar la clave unica
    hash = self.funcion_hash(value);
    if self.table[hash] is None:
      return None
    else:
      return hex(id(self.table[hash]))
```
